```python
# Nacti dva datove sety sttudentu do oddelenych pandas DataFrame a pomoci concat je spoj do jednoho setu.
import pandas as pd

jmena = pd.read_csv("jmena_2.csv")
s1 = pd.read_csv("studenti1.csv")
s2 = pd.read_csv("studenti2.csv")

studenti_all = pd.concat([s1, s2], ignore_index=True)
studenti_all.shape[0]
```

```python
# Pokud studentovi chybí ročník, znamená to, že již nestuduje. Pokud mu chybí číslo skupiny, znamená to, že jde o dálkového studenta.
# Kolik studentů v datovém setu již nestuduje a kolik jsou dálkoví studenti?
absolventi = (studenti_all[studenti_all['ročník'].isnull()])
print(absolventi.head(20))
absolventi.shape[0]
```

```python
dalkari = (studenti_all[studenti_all['kruh'].isnull()])
print(dalkari.head(20))
dalkari.shape[0]
```

```python
# Vyčisti data od studentů, kteří nestudují nebo studují jen dálkově. Nadále budeme pracovat pouze s prezenčními studenty.
prezencni_studenti = studenti_all.dropna()
prezencni_studenti.shape[0]
```

```python
# Zjisti, kolik prezenčních studentů je v každém z oborů.
pocet_studentu_oboru = prezencni_studenti['obor'].value_counts()
print(pocet_studentu_oboru)
```

```python
# Zjisti průměrný prospěch studentů v každém oboru.
prumerny_prospech = prezencni_studenti.groupby("obor")["prospěch"].mean()
print(prumerny_prospech)
```
