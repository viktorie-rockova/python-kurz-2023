```python
# Nacti dva datove sety sttudentu do oddelenych pandas DataFrame a pomoci concat je spoj do jednoho setu.
import pandas as pd

jmena = pd.read_csv("jmena_2.csv")
s1 = pd.read_csv("studenti1.csv")
s2 = pd.read_csv("studenti2.csv")

studenti_all = pd.concat([s1, s2], ignore_index=True)
studenti_all.shape[0]
```
