```python
# How to connect 2 conditions
lidnate_evropske_staty = staty[(staty["population"] > 20_000_000) & (staty["region"] == "Europe")]
lidnate_evropske_staty
```

```python
# Returns rows where either of the conditions are met
lidnate_evropske_staty = staty[(staty["population"] > 20_000_000) | (staty["region"] == "Europe")]
lidnate_evropske_staty
```
