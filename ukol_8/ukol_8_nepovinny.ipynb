```python
# Využij metody sort_values() (a libovolných dalších metod), ke zjištění následujících informací:

# Adopce kterých zvířat je nejdražší?
# Které zvíře je alfabeticky poslední v češtině? Které v angličtině?
# Bonus můžeš doplnit o nějaké vlastní pozorování.
```

```python
import pandas as pd

adopce = pd.read_csv("adopce-zvirat.txt", sep=';')
adopce.head
```

```python
# Adopce kterých zvířat je nejdražší?
pd.set_option("display.max_rows", 500)

nejdrazsi_zvire = adopce.sort_values(by='cena', ascending=False)

nejdrazsi_zvire
```

```python
# Které zvíře je alfabeticky poslední v češtině? Které v angličtině?

alfabeticky_posledni_cz = adopce.sort_values(by='nazev_cz', ascending=False)
alfabeticky_posledni_cz
```

```python
alfabeticky_posledni_en = adopce.sort_values(by='nazev_en', ascending=False)
alfabeticky_posledni_en
```
